```python
import numpy as np 
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')

df = pd.read_csv('TeenMental_health.csv')
df.head()
```

```python
age = df['age'].value_counts()
sns.barplot(x=age.index , y = age.values)
plt.xlabel('daily_social_media_hours')
plt.ylabel('strees_level')
plt.title('phone addication')
plt.show()
```

```python
sns.countplot(data=df, x='academic_performance', hue = 'physical_activity' )
plt.show()
```

```python
sns.violinplot( x = 'addication_level' , y = 'anxiety_level' , data = df ) ,
sns.swarmplot( x =  'addication_level'  ,  y = 'anxiety_level' , data = df , color ='black') 
plt.show()
```
